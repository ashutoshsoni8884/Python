```python
#merging twoo dataframe

player =['Play1','Play2','Play5']
Points=[1,2,4]
Title =['Game1','Game2','Game3']
df1 = pd.DataFrame({'Player':player,'Points':Points,'Title':Title})
df1
```

```python
player =['Play1','Play2','Play3']
Sports =['cric','football','basketball']
Title =['Game1','Game2','Game3']
df2 = pd.DataFrame({'Player':player,'Sports':Sports,'Title':Title})
df2
```

```python
df1.merge(df2, on='Player', how ='inner')
```

```python
#Left merge
df1.merge(df2, on='Player', how ='left')
```

```python
#Right merge
df1.merge(df2, on='Player', how ='right')
```

```python
#outer merge
df1.merge(df2, on='Player', how ='outer')
```

```python
#JOINS
#inner join
import pandas as pd
import numpy as np
player =['Play1','Play2','Play5']
Points=[1,2,4]
Titles =['Game1','Game2','Game3']
df3 = pd.DataFrame({'Player':player,'Points':Points,'Title':Titles}, index=['a','b','c'])
df3
```

```python
player =['Play1','Play2','Play3']
Sports =['cric','football','basketball']
Titles =['Game1','Game2','Game3']
df4 = pd.DataFrame({'Players':player,'Sports':Sports,'Titles':Titles}, index =['b','c','d'])
df4
```

```python
df3.join(df4, how ='inner')
```

```python
#left join
df3.join(df4, how ='left')
```

```python
#outer join
df3.join(df4, how ='outer')
```

```python
#concatenate using pandas
pd.concat([df3,df4])
```
